```python
import cv2

cap = cv2.VideoCapture("Lalu_lintas.mp4")

deteksi = cv2.createBackgroundSubtractorMOG2()

while True:
    ret, frame = cap.read()
    
    #untuk tambahan dari progres pertama
    #yaitu menambahkan frame/pembatas pada deteksi mobil, agar mendeteksi mobil dibagian yang sudah ditentukan
    area = frame [350:1120, 450:1000]
    
    objek = deteksi.apply(area)
    
    contours, _ = cv2.findContours(objek, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        luas = cv2.contourArea(cnt)
        if luas > 800:
            cv2.drawContours(area, (cnt), -1, (0,255,0), 3)
            
            #lalu menambahkan pendeteksi dengan contour yang berbentuk rectangle, agar deteksi objek lebih akurat
            x, y, w, h = cv2.boundingRect(cnt)
            cv2.rectangle(area, (x,y),(x+w, y+h), (0, 255, 0), 3)
            
    
    cv2.imshow("Lalu_lintas", frame)
    cv2.imshow("Objek", objek)
    cv2.imshow("Area", area)
    
    key = cv2.waitKey(30)
    if key == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
